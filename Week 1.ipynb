```python
from bs4 import BeautifulSoup
```

```python
url = "https://www.amazon.com/Fire-Fury-Inside-Trump-White/product-reviews/1250158060/ref=cm_cr_arp_d_viewopt_sr?ie=UTF8&reviewerType=all_reviews&pageNumber=1&filterByStar=three_star"
```

```python
import urllib2
```

```python
page = urllib2.urlopen(url)
```

```python
soup = BeautifulSoup(page, 'html5lib')
```

```python
print soup
```

```python
reviews = soup.findAll("span",{"data-hook":"review-body"})
```

print all the h1s

```python
print reviews
```
